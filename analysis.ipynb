```python
import pandas as pd
import re
```

```python
def remove_parentheses(filter_name):
    return re.sub(r'\(.*\)', '', filter_name).strip()
```

```python

excel_data = pd.ExcelFile('results.xlsx')

results = excel_data.parse('results')

results['Filter'] = results['Filter'].apply(remove_parentheses)

summary_stats = results.groupby('Filter').agg({
    'PSNR Gaussian': ['mean', 'std'],
    'PSNR Impulsive': ['mean', 'std'],
    'SSIM Gaussian': ['mean', 'std'],
    'SSIM Impulsive': ['mean', 'std'],
}).reset_index()

summary_stats.columns = ['Filter', 'PSNR Gaussian Mean', 'PSNR Gaussian Std',
                         'PSNR Impulsive Mean', 'PSNR Impulsive Std',
                         'SSIM Gaussian Mean', 'SSIM Gaussian Std',
                         'SSIM Impulsive Mean', 'SSIM Impulsive Std']

summary_stats
```

```python
excel_data = pd.ExcelFile('results_combined_noise.xlsx')

results = excel_data.parse('results_combined_noise')

results['Filter'] = results['Filter'].apply(remove_parentheses)

summary_stats = results.groupby('Filter').agg({
    'PSNR Gaussian + Impulsive': ['mean', 'std'],
    'SSIM Gaussian + Impulsive': ['mean', 'std']
}).reset_index()

summary_stats.columns = ['Filter', 'PSNR Gaussian + Impulsive Mean', 'PSNR Gaussian + Impulsive Std',
                         'SSIM Gaussian + Impulsive Mean', 'SSIM Gaussian + Impulsive Std']

summary_stats
```
